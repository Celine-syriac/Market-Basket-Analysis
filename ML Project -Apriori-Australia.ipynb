```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df=pd.read_excel(r"C:\Users\user\Desktop\Introtallent\Python\data\Online Retail.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df['Country'].unique()
```

```python
df['Description']=df['Description'].str.strip()
```

```python
df.isnull().sum()
```

```python
#looking at sales for 
basket=(df[df['Country']=='Australia']
       .groupby(['InvoiceNo','Description'])['Quantity'].sum()
       .unstack()
       .reset_index().fillna(0)
       .set_index('InvoiceNo'))
```

```python
basket.shape
```

```python
def replace_quantity(x):
    if x>=1:
        return 1
    else:
        return 0
    
basket_sets=basket.applymap(replace_quantity)
```

```python
basket_sets.to_excel(r"C:\Users\user\Desktop\Introtallent\Python\basket_data_Australia_.xlsx")
```

```python
basket_sets.drop('POSTAGE',inplace=True,axis=1)
```

```python
frequent_itemsets=apriori(basket_sets,min_support=0.07,use_colnames=True)
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
rules
```

```python
rules[(rules['lift']>=6)&(rules['confidence']>=0.8)]
```

```python
#export association rules to excel
rules.to_excel(r"C:\Users\user\Desktop\Introtallent\Python\output_.xlsx")
```
